```python
from IPython.display import Video
import moviepy.editor as moviepy
```

```python
import torch

# Check cuda version and availability
print(f"CUDA version: {torch.version.cuda}")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
  print(f"GPU name: {torch.cuda.get_device_name(0)}")
```

```python
# View all settings
print(f"YOLO-settings: {settings}")
```

```python
# Load a pretrained model
#
model_path = "weights/yolo11x"

# Load model
model = YOLO(model_path)
```

```python
# Display model information (optional)
print(model.info())
```

```python
# Data source
source = "/content/sample_data/fish002_640p.mp4"

# Output sub-directory
output_dir = 'fish002_YOLO11x_ByteTrack'

# Perform tracking with the model
# results = model.track(source, show=False, save=True, name=output_dir, device=0, tracker="bytetrack.yaml")
results = model.track(source, show=False, save=True, tracker="bytetrack.yaml", device=0)
```

```python
# Convert avi file to mp4

# input and output path
input_video_path = "/content/runs/detect/track2/fish002_640p.avi"
output_video_path = "/content/runs/detect/track2/fish002_640p_YOLO11x.mp4"

# Conversion
clip = moviepy.VideoFileClip(input_video_path)
clip.write_videofile(output_video_path)
```

```python
# Display output video as embeded
# output_video_path = "/content/runs/detect/track2/fish002_640p_YOLO11x.mp4"
Video(output_video_path, embed=True)
```
